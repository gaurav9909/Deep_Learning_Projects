```python
import tkinter as tk
from tkinter import ttk
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
import pandas as pd
from heapq import nlargest

def summarize_text():
    text = text_input.get("1.0", tk.END)
    if not text:
        result_label.config(text="Please enter text to summarize.")
        return

    # Your existing summarization code goes here
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct and token.text != '\n']
    tokens1 = []
    stopwords = list(STOP_WORDS)
    allowed_pos = ['NOUN', 'PROPN', 'ADJ', 'ADV']
    for token in doc:
        if token.text in stopwords or token.text in punctuation:
            continue
        if token.pos_ in allowed_pos:
            tokens1.append(token.text)

    word_freq = Counter(tokens)
    max_freq = max(word_freq.values())
    for word in word_freq.keys():
        word_freq[word] = word_freq[word] / max_freq

    sent_token = [sent.text for sent in doc.sents]
    sent_score = {}
    for sent in sent_token:
        for word in sent.split(' '):
            if word.lower() in word_freq.keys():
                if sent not in sent_score.keys():
                    sent_score[sent] = word_freq[word]
                else:
                    sent_score[sent] += word_freq[word]

    num_sentences = 5
    n = nlargest(num_sentences, sent_score, key=sent_score.get)
    summary = " ".join(n)

    result_label.config(text=summary)

# Create the main window
root = tk.Tk()
root.title("Text Summarizer")

# Create and place widgets
text_input = tk.Text(root, wrap=tk.WORD)
text_input.pack(padx=10, pady=10)

summarize_button = ttk.Button(root, text="Summarize", command=summarize_text)
summarize_button.pack(pady=5)

result_label = ttk.Label(root, text="", wraplength=500)
result_label.pack(padx=10, pady=10)

root.mainloop()
```
